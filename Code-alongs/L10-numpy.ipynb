```python
number_list = [1, 2, 3] # ordinary list
number_list = number_list*2 #double the list
print(number_list) #run the list. it is now 1,2,3,1,2,3

# in a list we can place different data types
mixed_list = [1, "ett", True, 2.3] # dynamic typing in Python
#You can multiply this by * 2 as well. 
# each element in the list is an object, it has information on what type it is
# this requers some overhead, which slows down the program.
# note that when we have one data type, it is unnecessary to have this overhead,
print(mixed_list)
#numpy can solve this issue
```

```python
# in numpy we have statically typed
import numpy as np

# create an array from a list
vector1 = np.array([2,5,1])
print(vector1)
vector2 = vector1*2
print(vector2)

print(f"{vector1+2=}") #Every number adds

#indexing
print(f"{vector1[2]=}")
print(f"{vector1[-1]=}")

#slice
print(f"{vector1[1:]=}")

#change value in array, index
vector1[1] = 1337
print(vector1)

# can't append to an array - can't change its size
# can't have different types in array

#sort
vector1.sort()
print(f"{vector1=}")

#sum
print(f"{vector1.sum()=}") # 2 + 1337 + 1 = 1340
print(f"{vector1.mean()=:.2f}")

```

```python
ones_vector = np.ones((10))
print(f"{ones_vector}")

#2D array of shape 3,3
matrix_3x3 = np.ones((3,3))

print(f"{matrix_3x3=}")


#We want to acess the second index in the third row:
# [row, index] - counts from 0
matrix_3x3[2,1] = 10

matrix_3x3
```

```python
print(matrix_3x3.shape)

# slicing operator to get row
print(f"{matrix_3x3[2,:]=}")

#slicing operator to get a column
print(f"{matrix_3x3[2,:]=}")

# add 2 vectors
matrix_3x3[2,:] + matrix_3x3[:,1]
```
