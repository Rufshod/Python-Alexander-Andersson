# Functions

```python
#defina a function called cuber, it takes in 1 parameter
def cuber(x):
    return x**3

#calls the function cuber using 1 argument, in this case 5
print(f"{cuber(5) = }")

cubes = [cuber(i) for i in range(10)]
print(cubes)
```

```python
import matplotlib.pyplot as plt

plt.plot(cubes);
```

```python
# 3 parameters
def my_mean(number1, number2, number3):
    print(f"{number1 = }, {number2 = }, {number3 = }")
    return (number1+number2+number3)/3
# 3 arguments
print(my_mean(1,3,6))
print(my_mean(4,7,10))

double_mean = my_mean(1,2,3)*2
print(double_mean)
```

## Default values

```python
#You can insert a default value, so if the function lacks argument it will be the default value
#in this example number_of_greetings does not have a default value, but name and age has "buddy" and age 0 respectivly.

def hello(number_of_greetings, name = "buddy", age = 0):
    if float(age) < 0:
        print("Ange en ålder större än = 0")
        return #If we do not add return here the program will continue to print if the age is less than 0
    for i in range(number_of_greetings):
        print(f"Hej {name}, du är {age} år gammal")
    #note, no return → void function

hello(2)

hello(1, "Bella", -10) # ange en ålder is printed bc of the return in the if statement.

# age and name are keyword arguments 
# number of greetings is a positional argument
hello(1, age = 13)

```

```python
def word_counter(text = ""):
    word_list = text.split()
    return len(word_list)

quote = "No I am your father - Anakin Skywalker"
word_counter(quote)
```

## Arbitrary arguments - *args

```python
def my_mean2(*numbers): #* before numbers means that we can add as many positional arguments as we want.
    numsum = 0
    for number in numbers:
        numsum += number
    return numsum/len(numbers)

my_mean2(1,2,3,4,5,6,7,8,9)
```

## Lamda

```python
cuber = lambda x: x**3

cuber(3)
```

## Draw lines

```python
def draw_lines(x_list, k=1, m=1, title = ""):
    y_list = [k*x+m for x in x_list]

    plt.plot(x_list, y_list)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title(title)

# TODO find intersection point and plot it

x = list(range(-10,10))

draw_lines(x, title = "Lines")
draw_lines(x, k=-2, m=-1, title = "Lines")

print(x)
```
