```python
import matplotlib.pyplot as plt
import numpy as np

# array from -10 to 10 with 100 numbers equally spaced between 
x = np.linspace(-10,10, 101)
f = lambda x: np.cos(x) #x is not the same x as above. 
# same as lambda â†“
# def f(x)
#   return np.cos()

print(x[2])
f(2)
```

## MATLAB style interface

```python
plt.title("Plot goes ooaAaoaAaoaAaoo ")
plt.grid(False)
plt.box(False)
plt.plot(x, f(x), 'o--', c="r");
```

```python
x = np.linspace(-10,10, 200)

plt.figure(figsize=(8,4), dpi = 150)
plt.title("cos(x)")
plt.grid()
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x, f(x), '--', c="darkblue");
```

## OOP Approach

```python

#instatiated fig, ax from figure and axes.
fig, ax = plt.figure(figsize = (8,4)), plt.axes()
ax.plot(x, f(x)) # ax object as a plot method
ax.set(xlabel = "x", ylabel ="y", title = "$\cos{x}$", xlim = (-5,5), ylim = (-1.2,1.2))
fig.savefig("../assets/cos(x).png", dpi=200)
```

## Subplots

```python
# MATLAB style

plt.figure(figsize=(16,5))
plt.subplot(1,2,1) # (rows, columns, panel number)
plt.title("cos(x) sin(x)")
plt.plot(x, f(x))
plt.plot(x, np.sin(x))

plt.subplot(1,2,2) # new plot because the panelnumber is new
plt.title("cos(x)")
plt.plot(x, f(x))

plt.suptitle("Trigonometric functions", fontweight = "bold")
```

```python
# OOP style

fig, ax = plt.subplots(1,2, figsize=(16,5))

ax[0].plot(x,f(x))
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.sin(x))

ax[0].set(title = "cos(x) & sin(x)" )

fig.suptitle("Big Title", fontsize = 25)
```
