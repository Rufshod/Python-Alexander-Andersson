## Example - birds and oil

There is an oul leakage causingf the bird population in an island to 1/2 in each year. From start there were 80000 birds, how many years does it take for it to have 1/10 remaning

````python
birds = 80000
years = 0
while birds >= 8000:
    birds/=2
    years+=1
    
    print(f"There are {birds:.0f} remaining year {years}")
    if birds <= 8000:
        print(f"The year that there are only 1/10 birds remaining is {years - 1} ")
````

````python
#Teachers solution
tbirds, tyears = 80000, 0 #tuple unpacking.
limit = tbirds/10
while tbirds >= limit:
    tbirds /= 2
    tyears += 1
    print(f"Year {tyears}: nuber of birds {tbirds:.0f} left")
````

## Example Milk

1L of milk has 1 500 000 bacterias in room temperature
Bacterias increases with 50% each hour in room temp
Milk gets soured after 10 000 000 bacterias
How many hours does it take until milk sours (whole hours)

````python
bacterias = 1500000
increase = 1.5 # 50% increase
sour = 10000000
hour = 0
print("With a 50% increase in bacteria,")
while bacterias <= sour:
    bacterias*=increase
    hour+=1
    
    print(f"During hour {hour} the bacteria level is at {bacterias:.0f}  ")
````

````python
#Teachers Solution
tbacterias = 1.5e6 # the same as 1.5 * 10^6
tfactor = 1.5
thours = 0
tsour = 1e7 # 10^7

tbacterias_list = [tbacterias]
print(tbacterias_list)

while tbacterias < tsour:
    tbacterias *= tfactor
    thours += 1

print(f"Takes {thours}h for milk to get soured") 

````

````python
#Teachers Solution2 with import
import matplotlib.pyplot as plt # as plt = we can use the module with .plt

tbacterias = 1.5e6 # the same as 1.5 * 10^6
tfactor = 1.5
thours = 0
tsour = 1e7 # 10^7

tbacterias_list = [tbacterias]
print(tbacterias_list)


while tbacterias < tsour:
    tbacterias *= tfactor
    thours += 1
    tbacterias_list.append(tbacterias)

print(tbacterias_list)
print(f"Takes {thours}h for milk to get soured") 

#This creates a range object with numbers: 0,1,2,...,hours
thours_list = range(0, thours+1)

plt.plot(thours_list, tbacterias_list, 'o-') #o is the dots, - is the line, -- gets dotted lines
plt.plot(thours_list, [tsour]*(thours+1))
plt.grid() #creates a grid
plt.xlabel("Hours")
plt.ylabel("Number of bacterias")
plt.title("Growth of bacterias in milk at room temperature")

````
