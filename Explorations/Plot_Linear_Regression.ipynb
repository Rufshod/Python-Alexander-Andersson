## Experimenting with matplot + linearregression

```python
import matplotlib.pyplot as plt
import numpy as np
x_list =[43, 21, 25, 42, 57, 59]
y_list = [99, 65, 79, 75, 87, 81]

plt.plot(x_list, y_list)
plt.scatter(x_list, y_list, color="red",marker="x")
plt.xlabel("Urban Density")
plt.ylabel("Pokemon Quantity")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

x = 1000 * np.random.random((100, 1))

y = 0.01 * x + 1.0 + np.random.normal(size=x.shape)

model = LinearRegression()
model.fit(x, y)

x_new = np.linspace(0, 1000, 1000)
y_new = model.predict(x_new[:, np.newaxis])

plt.figure(figsize=(10, 4))
ax = plt.axes()
ax.scatter(x, y)
ax.plot(x_new, y_new)

ax.set_xlabel('House Prices (Tusen kr)')
ax.set_ylabel('Antal Rum')

ax.axis('tight')

plt.show()
```
