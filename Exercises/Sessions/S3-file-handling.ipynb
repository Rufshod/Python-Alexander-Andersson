````python
with open ("../../Data/Provresultat.txt") as f_read: # no r in (), read is default
    line = f_read.readline() #gets one line from document
    #each line is stored in a list
    #each line is a string
    #each line is a string and has a \n in the end
    lines = f_read.readlines() #prints everything, as a list
    
    print(lines)

    #---------------------a) print out the text in the terminal --------------------------
    #loops through every line in lines and print out line for every element, ends with nothing ""
    for line in lines:
        print(line, end = "")
````

````python
# b) and c)
# saved as write complete 
with open ("../../Data/Provresultat.txt", "r") as f_read, open("../../Data/Provresultat_tmp.txt", "w") as f_write:
    lines = f_read.readlines()
    #add a new line to last index so that it prints correctly
    #print(lines[-1])

    lines[-1] += "\n"

    f_write.write("\n\nAlfabetisk ordning\n")
    #print(lines)

#Sorting :

    # returns None, it mutates (changes) the list
    #lines.sort()
    #print(lines)
# ------------------- b) ------------------- #
    #another way, returns a new list of sorted elements
    lines_sorted_alphabetically = sorted(lines)
    #print(lines_sorted_alphabetically)

    for line in lines_sorted_alphabetically:
        f_write.write(line)

    f_write.write("\n\nSorterat efter betyg\n\n")
# ------------------- c) ------------------- #
# need to split and find the score for each line
# put scores in grade order
    
    grades = list("ABCDEF")[::-1] # setting grades as list and reverses
    print(grades)


    for grade, limit in zip(grades, range(20, 71, 10)): #10 is to jump ten numbers evertime
        f_write.write(f"\nBetyg {grade}\n")

        for line in lines:
            #print(line.split()) #this gives a list for each line, splits at all whitespace, newline, space and tab
            print(line.split()[-1])


            point = int(line.split()[-1]) # type converts string to int
            # if statement requires test between same types

            if limit-10 <= point < limit:
                f_write.write(f"{line}") # type converts int to string so that we can write to txt file, can not write int to txt file.
````
